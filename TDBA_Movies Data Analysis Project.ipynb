## Data Wrangling

>We will load in the data, check for cleanliness, and then trim and clean our dataset prepareing it for analysis.

.



```python

df=df=pd.read_csv('tmdb_movies_data.csv')
df.head()
```



### Data Cleaning 

```python
# general information a bout our dataframe
df.info()
```

```python
# drop null values for genres column
df.drop(index=df[df['genres'].isna()].index,inplace=True)
```

```python
# drop null values for cast column

df.drop(index=df[df['cast'].isna()].index,inplace=True)
```

```python
# drop null values for director column

df.drop(index=df[df['director'].isna()].index,inplace=True)
```

```python
df.info()
```

#####  in the next steps solving the problem of genres column making each feature in a whole column


```python
df['genres']=df['genres'].str.split('|')
```

```python
# def a function to Exaggerate the lists then we use it to make a unique list of genres
def to_series(series):
    return pd.Series ([x for list_ in series for x in list_])
```

```python
# making a unique list of genres
unique_list=to_series(df['genres']).value_counts().index
```

```python
unique_list
```

```python
# make a copy of dataframe and  push the generes column_at the end of the new frame as new name "features "
movies=df.copy()
```

```python
features=df.pop("genres") # save our genres column into  another column 'features'
```

```python
movies["features"]=features
```

```python
movies.drop(columns=['genres'],inplace=True)
```

```python
movies.rename(columns={"features":"genres"},inplace=True)
```

```python
movies # our new data frame
```

```python
# def a function making a new dataframe each column contain only one feature using our unique list which previously created
def mask_genres(series,list_):
    dict_={}
    for i, item in enumerate (list_):
        dict_[item]=series.apply(lambda x: item in x)
    return pd.DataFrame(dict_)
```

```python
genres_frame = mask_genres(movies['genres'],unique_list)
genres_frame
```

```python
# replace true,false with 1,0
genres_frame.replace([True,False],[1,0],inplace=True)
```

```python
# create new frame with extra columns each genres in it's whole column in adjacent to the essential genres column
movies_genres=pd.concat([movies,genres_frame],axis='columns',sort=False)
```

```python
movies_genres.info()
```

### Question2 ###
what is the most Apperance actors in this data sets

```python
movies_genres.reset_index(inplace=True)
```

```python
actors = movies_genres["cast"]
actors = actors.str.split("|")
#actors
```

```python
act_dict={}
for actor in actors:
    for i in actor:
        if i not in act_dict:
            act_dict[i]=1
        else:
            
            act_dict[i]+=1 
        
#act_dict
```

```python
sorted_act_=sorted(act_dict.items(),key=itemgetter(1),reverse=True)
#sorted_act_
```

```python
x_list=[]
y_list=[]
colors=['blue','brown','green','red','purple','yellow','pink','grey','orange','black']
for i in sorted_act_[0:10]:
    x_list.append(i[0])
    y_list.append(i[1])
xticks_location=np.arange(len(x_list[0:10]))    
plt.figure(figsize=(18,5))    
plt.bar(x_list,y_list,width=0.5,color=colors)
plt.xticks(xticks_location,x_list,fontweight='bold')
plt.title('Actors with the most apperance',fontsize=18)
plt.ylabel('Number of apperance',fontsize=15)
plt.show()
```

the most apperance actor is **Robert De Niro** about 72 movies
